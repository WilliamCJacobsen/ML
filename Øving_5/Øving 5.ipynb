Visualise the feature space using a similar method to the one we used for the 20 newgroups dataset.

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Data/agaricus-lepiota.csv")

dataset.head()
```

```python
dataset_dummy = pd.get_dummies(dataset)
dataset_dummy.head()
```

```python
def plot_barchart(number_of_x, label_x, number_of_y, label_y ,title = "default"):
    # this is for plotting purpose
    index = np.arange(2)
    plt.bar([label_x, label_y], [number_of_x, number_of_y])
    plt.ylabel("amount", fontsize=10)
    plt.xticks(index, [label_x, label_y], fontsize=10, rotation=30)
    plt.title(title)
    plt.show()
```

```python
edible = len(dataset_dummy[dataset_dummy['edible_p'] == 1])
poision = len(dataset_dummy[:]) - edible

plot_barchart(edible, "edible", poision, "poision", title = "edible")
```

```python
yellow = len(dataset_dummy[dataset_dummy['edible_p'] == 1] [dataset_dummy['cap-color_y'] == 1])
brown = len(dataset_dummy[dataset_dummy['edible_p'] == 1] [dataset_dummy['cap-color_n'] == 1])
white = len(dataset_dummy[dataset_dummy['edible_p'] == 1] [dataset_dummy['cap-color_w'] == 1])
gray = len(dataset_dummy[dataset_dummy['edible_p'] == 1] [dataset_dummy['cap-color_g'] == 1])

index = np.arange(4)
plt.bar(["yellow", "brown", 'white', 'gray'], [yellow, brown, white, gray ])
plt.ylabel("amount", fontsize=10)
plt.xticks(index, ["yellow", "brown", 'white', 'gray'], fontsize=10, rotation=30)
plt.title("Edible colours")
plt.show()

```
