```python
import keras
from keras.utils import to_categorical
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras import layers
from tensorflow import keras
```

```python
x_data = ["rat  ", "house", "hat  ", "cat  ", "plane", "zack "]
y_data = ["üêÄ ","üè† " ,"üé© ", "üêà ", "üõ© ", "ü§ê"]
int_char = dict((i, c) for (i,c) in enumerate("".join(set(x_data))))
dictonary = dict((c, i) for (i,c) in enumerate("".join(set(x_data))))
y_len = len(y_data)
print(dictonary)
```

```python
def to_string(*values):
    return [ list(keras.utils.to_categorical(list(map(lambda x: dictonary[x], list(string))), 30)) for string in values ]
```

```python
y_train = np.eye(6)
x_train = np.array([ to_string(word) for word in x_data ]).reshape(6,5,30)
print(np.array(x_train).shape)
print(np.array(y_train).shape)
```

```python
units_in_LSTM = 128
epochs = 500
(_,_,encoding_size) = np.array(x_train).shape

model = keras.Sequential([
    keras.layers.LSTM(units_in_LSTM, input_shape=(None, encoding_size), return_sequences=False),
    keras.layers.Dense(y_len , activation='softmax')
])

optimizer = tf.train.RMSPropOptimizer(learning_rate = 0.05)
loss = tf.keras.losses.categorical_crossentropy

model.compile(loss= loss, optimizer=optimizer)

model.summary()
```

```python
model.fit(
    x_train, 
    y_train, 
    batch_size=1, 
    steps_per_epoch=1,
    epochs=epochs, 
    verbose=True,
)
```

```python
def test(str):
    ret = model.predict([to_string(str)])[-1].argmax()
    print(ret)
    return y_data[ret]

test("ratss")
```
